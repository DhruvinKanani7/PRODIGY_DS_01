```python
# Load the CSV file, skipping the initial rows if necessary
df = pd.read_csv('/content/sample_data/API_SP.POP.TOTL_DS2_en_csv_v2_900.csv', encoding='utf-8', skiprows=4)

# Drop any columns that are not part of the data
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

#country
chosen_country = 'India'
# Filter the data for the chosen country
country_data = df[df['Country Name'] == chosen_country]

# Ensure that the columns for years are properly identified
years = country_data.columns[4:].astype(int)  # Adjust the index if necessary
population = country_data.iloc[0, 4:].values.astype(float)  # Adjust the index if necessary

# Plot the data
plt.figure(figsize=(10, 5))
plt.bar(years, population, color='red')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title(f'Population of {chosen_country} Over the Years')
plt.show()
```

```python
#histogram
plt.hist(population, bins=30, color='r', edgecolor='b')
plt.show()
```
